```python
import cv2
import numpy as np
import face_recognition
import os
from datetime import datetime

# Path to the images directory
path = 'images'
images = []
classNames = []

# Get list of all files in the directory
myList = os.listdir(path)
print("Images found:", myList)

# Load images and extract class names
for cl in myList:
    curImg = cv2.imread(f'{path}/{cl}')
    if curImg is not None:  # Check if the image is loaded properly
        images.append(curImg)
        classNames.append(os.path.splitext(cl)[0])
print("Class Names:", classNames)

# Log file for training and testing details
log_file = 'log.txt'

def log_message(message):
    with open(log_file, 'a') as log:
        log.write(f"{datetime.now()} - {message}\n")

# Function to find encodings for all images
def findEncodings(images):
    encodeList = []
    for i, img in enumerate(images):
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        encodes = face_recognition.face_encodings(img)
        if encodes:  # Ensure that at least one face is found
            encodeList.append(encodes[0])
            log_message(f"Image {i+1}/{len(images)} encoded successfully.")
        else:
            log_message(f"Image {i+1}/{len(images)} has no detectable faces.")
    return encodeList

# Function to mark attendance
def markAttendance(name):
    with open('ATTENDANCE.csv', 'r+') as f:
        myDataList = f.readlines()
        nameList = [line.split(',')[0] for line in myDataList]
        
        if name not in nameList:
            now = datetime.now()
            dtString = now.strftime('%H:%M:%S')
            f.write(f'\n{name},{dtString}')
            log_message(f"Attendance marked for {name} at {dtString}.")

# Get encodings for all known images
encodeListKnown = findEncodings(images)
print('Encoding Complete')

# Initialize webcam
cap = cv2.VideoCapture(0)

while True:
    success, img = cap.read()
    
    # Check if the frame is captured successfully
    if not success:
        print("Failed to capture image")
        log_message("Failed to capture image from webcam.")
        continue
    
    # Reduce size of the image for faster processing
    imgS = cv2.resize(img, (0, 0), None, 0.25, 0.25)
    imgS = cv2.cvtColor(imgS, cv2.COLOR_BGR2RGB)

    # Find all faces and their encodings in the current frame of video
    facesCurFrame = face_recognition.face_locations(imgS)
    encodesCurFrame = face_recognition.face_encodings(imgS, facesCurFrame)
    log_message(f"Detected {len(facesCurFrame)} faces in the current frame.")

    # Loop through each face found in the frame
    for encodeFace, faceLoc in zip(encodesCurFrame, facesCurFrame):
        matches = face_recognition.compare_faces(encodeListKnown, encodeFace)
        faceDis = face_recognition.face_distance(encodeListKnown, encodeFace)
        
        # Find the best match for the recognized face
        matchIndex = np.argmin(faceDis)

        if matches[matchIndex]:
            name = classNames[matchIndex].upper()
            y1, x2, y2, x1 = faceLoc
            y1, x2, y2, x1 = y1 * 4, x2 * 4, y2 * 4, x1 * 4
            
            # Draw a rectangle around the face
            cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.rectangle(img, (x1, y2 - 35), (x2, y2), (0, 255, 0), cv2.FILLED)
            cv2.putText(img, name, (x1 + 6, y2 - 6), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 255, 255), 2)
            
            # Mark attendance
            markAttendance(name)
            log_message(f"Match found for {name}.")

    # Display the video feed
    cv2.imshow('Webcam', img)
    cv2.waitKey(1)
```
